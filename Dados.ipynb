```python
import random



def rolar_dados(quantidade_de_dados,numero_de_faces):
    total=[]    
    for rolagem in range(quantidade_de_dados):
        resultado_do_dado=random.randint(1,numero_de_faces)
        total.append(resultado_do_dado)
    
    
    sum(total)
    

for rolagem in range(100):
    rolar_dados(6,10)
    
    
    

```
